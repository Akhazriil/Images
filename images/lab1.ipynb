```python
import numpy as np

def create_list(n: int) -> np.ndarray:
    np.random.seed(314)
    x = np.random.uniform(0, 100, n)
    x = np.round(x, 2)
    x = np.where( x > 90, x - 10, x)
    x = np.where( x < 10, x + 10, x)
    x_mean = x.mean()
    x_min = x.min()
    x_max = x.max()
    x_sum = int(x_mean + x_min + x_max)
    return x_sum

print(create_list(10))
```

```python
import numpy as np

def matrix_processing(m, n):
    matrix = np.zeros((m, n), dtype=int)
    last_num = (min(m, n) + 1) // 2
    for i in range(last_num):
        number = 1 + i
        matrix[i, i:n-i] = number
        matrix[m - 1 - i, i:n-i] = number
        matrix[i+1:m - 1 - i, i] = number
        matrix[i+1:m - 1 - i, n - i - 1] = number
    sum_matrix = np.sum(matrix)

    return sum_matrix

print(matrix_processing(4,5))
```

```python
import numpy as np

def matrix_processing(m, n):
    matrix = np.zeros((m, n), dtype=int)
    layers = (min(m, n) + 1) // 2
    
    for i in range(layers):
        value = i + 1
        matrix[i:m-i, i:n-i] = value
    return np.sum(matrix)

print(matrix_processing(4,5))
```

```python
import numpy as np

def temperature_prosessing(n):
    np.random.seed(314)
    x = np.random.uniform(15, 33, 30)
    y = np.random.uniform(15, 33, 30)
    concatenated_arr = np.concatenate([x[x > n],y[y > n]])
    ar_mean = np.mean(concatenated_arr)
    ar_min = np.min(concatenated_arr)
    ar_max = np.max(concatenated_arr)
    ar_std = np.std(concatenated_arr)
    ar_num = len(concatenated_arr)
    sum_concatenated_arr = int(ar_num + ar_mean + ar_min + ar_std + ar_max)

    return sum_concatenated_arr

print(temperature_prosessing(24))
```
